```python
import pandas as pd 
import numpy as np 
from matplotlib import pyplot as plt 
import seaborn as sns 
import time 
sns.set_style('darkgrid')
x = np.linspace(-10, 10, 200)
A, B, C = -2, 3, 4
learning_rate = 0.3

for i in range(3):
    y = (A*x - C)/B
    plt.plot(x, y)
    plt.pause(0.01)
    remove = np.random.choice([-1, 1])*learning_rate
    A, B, C = A*(1 - remove), B*(1 - remove), C*(1 - remove)

plt.show()
```
